```python
# 1
import japanize_matplotlib
import matplotlib.pyplot as plt
import random

X_1000 = [random.random() for _ in range(1000)]
X_10000 = [random.random() for _ in range(10000)]
X_100000 = [random.random() for _ in range(100000)]


def show_graph(X, title, bins):
    plt.figure(figsize=(8, 6))
    plt.title(title)
    plt.xlabel("")
    plt.ylabel("")

    plt.hist(X, bins=bins, color="blue", rwidth=1.0)
    plt.grid()
    plt.show()

show_graph(X_1000, "1000", 30)
show_graph(X_10000, "10000", 30)
show_graph(X_100000, "100000", 30)
```

```python
# 3
def show_scatter_graph(X, Y, label, x_label, y_label, color, title):
    plt.figure(figsize=(5, 4))
    plt.title(title)
    plt.scatter(X, Y, label=label, color=color)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.grid()
    plt.show()

X_100 = [random.uniform(-5, 5) for _ in range(100)]
Y_100 = [random.uniform(-5, 5) for _ in range(100)]

X_1000 = [random.uniform(-5, 5) for _ in range(1000)]
Y_1000 = [random.uniform(-5, 5) for _ in range(1000)]

X_10000 = [random.uniform(-5, 5) for _ in range(10000)]
Y_10000 = [random.uniform(-5, 5) for _ in range(10000)]

show_scatter_graph(X_100, Y_100, "-5 <= x <= 5, -5 <= y <= 5", "x", "y", "black", "-5 <= x <= 5, -5 <= y <= 5100")
show_scatter_graph(X_1000, Y_1000, "-5 <= x <= 5, -5 <= y <= 5", "x", "y", "black", "-5 <= x <= 5, -5 <= y <= 51000")
show_scatter_graph(X_10000, Y_10000, "-5 <= x <= 5, -5 <= y <= 5", "x", "y", "black", "-5 <= x <= 5, -5 <= y <= 510000")
```

```python
# 4
import math

X = [random.uniform(-1, 1) for _ in range(10000)]
Y = [math.cos(2 * math.pi * x) + random.uniform(-0.5, 0.5) for x in X]

show_scatter_graph(X, Y, "y = cos(2x) + noise", "x", "y", "black", "y = cos(2x)")
```

```python
# 5
X = ["X{}".format(x) for x in range(10)]
Y = [5 * random.random() for _ in range(10)]

XY = [[x, y] for x, y in zip(X, Y)]
XY.sort(key=lambda x: x[1],  reverse=False)

plt.figure(figsize=(5, 4))
plt.title("")
plt.barh([xy[0] for xy in XY],[xy[1] for xy in XY])
plt.xlabel("")
plt.ylabel("")
plt.grid()
plt.show()
```

```python
# 6
X = ["{}".format(x) for x in range(2020, 2024)]
Y = [random.uniform(0, 100) for _ in range(4)]

XY = [[x, y] for x, y in zip(X, Y)]
XY.sort(key=lambda x: x[1], reverse=True)

plt.figure(figsize=(5, 4))
plt.title("202020231")
plt.bar([xy[0] for xy in XY],[xy[1] for xy in XY])
plt.xlabel("")
plt.ylabel("")
plt.grid()
plt.show()
```

```python
# 7
product = ["Vanilla", "Strawberry", "Green tea", "Chocolate", "Mint"]
sales = [random.randint(0, 200) for _ in range(5)]

pie_colors = ["linen", "red", "forestgreen", "sienna", "mediumspringgreen"]
plt.rcParams['font.size'] = 16.0
plt.pie(sales, labels=product, counterclock=True, colors=pie_colors, wedgeprops={'linewidth': 1, 'edgecolor': "black"})
plt.show()
```
