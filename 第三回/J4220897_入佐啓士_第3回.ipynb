```python
# 課題1
import japanize_matplotlib
import matplotlib.pyplot as plt
import random

X_1000 = [random.random() for _ in range(1000)]
X_10000 = [random.random() for _ in range(10000)]
X_100000 = [random.random() for _ in range(100000)]


def show_graph(X, title, bins):
    plt.figure(figsize=(8, 6))
    plt.title(title)
    plt.xlabel("乱数の値")
    plt.ylabel("個数")

    plt.hist(X, bins=bins, color="blue", rwidth=1.0)
    plt.grid()
    plt.show()

show_graph(X_1000, "1000個の一様乱数のヒストグラム", 30)
show_graph(X_10000, "10000個の一様乱数のヒストグラム", 30)
show_graph(X_100000, "100000個の一様乱数のヒストグラム", 30)
```

```python
# 課題1の考察
"""
乱数の値を増やすと、ヒストグラムのばらつきが均一に近づいている。これは試行回数が増えることで乱数が等しく散らばるからだと考える。
"""
```

```python
#課題2
X_dice = [random.randint(1, 6) for _ in range(1000)]

plt.figure(figsize=(8, 6))
plt.title("サイコロ1000回の実行結果のヒストグラム")
plt.xlabel("サイコロの目")
plt.ylabel("出現回数")

plt.hist(X_dice, bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.6], color="blue", rwidth=0.5)
plt.grid()
plt.show()
```

```python
# 課題3
def show_scatter_graph(X, Y, label, x_label, y_label, color, title):
    plt.figure(figsize=(5, 4))
    plt.title(title)
    plt.scatter(X, Y, label=label, color=color)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.grid()
    plt.show()

X_100 = [random.uniform(-5, 5) for _ in range(100)]
Y_100 = [random.uniform(-5, 5) for _ in range(100)]

X_1000 = [random.uniform(-5, 5) for _ in range(1000)]
Y_1000 = [random.uniform(-5, 5) for _ in range(1000)]

X_10000 = [random.uniform(-5, 5) for _ in range(10000)]
Y_10000 = [random.uniform(-5, 5) for _ in range(10000)]

show_scatter_graph(X_100, Y_100, "-5 <= x <= 5, -5 <= y <= 5", "x", "y", "black", "-5 <= x <= 5, -5 <= y <= 5のランダムな100個の散布図")
show_scatter_graph(X_1000, Y_1000, "-5 <= x <= 5, -5 <= y <= 5", "x", "y", "black", "-5 <= x <= 5, -5 <= y <= 5のランダムな1000個の散布図")
show_scatter_graph(X_10000, Y_10000, "-5 <= x <= 5, -5 <= y <= 5", "x", "y", "black", "-5 <= x <= 5, -5 <= y <= 5のランダムな10000個の散布図")
```

```python
# 課題4
import math

X = [random.uniform(-1, 1) for _ in range(10000)]
Y = [math.cos(2 * math.pi * x) + random.uniform(-0.5, 0.5) for x in X]

show_scatter_graph(X, Y, "y = cos(2πx) + noise", "x", "y", "black", "ノイズを含んだ関数y = cos(2πx)の散布図")
```

```python
# 課題5
X = ["X{}".format(x) for x in range(10)]
Y = [5 * random.random() for _ in range(10)]

XY = [[x, y] for x, y in zip(X, Y)]
XY.sort(key=lambda x: x[1],  reverse=False)

plt.figure(figsize=(5, 4))
plt.title("ランダムな数値の棒グラフ")
plt.barh([xy[0] for xy in XY],[xy[1] for xy in XY])
plt.xlabel("ラベル名")
plt.ylabel("ラベルの値")
plt.grid()
plt.show()
```

```python
# 課題6
X = ["{}年".format(x) for x in range(2020, 2024)]
Y = [random.uniform(0, 100) for _ in range(4)]

XY = [[x, y] for x, y in zip(X, Y)]
XY.sort(key=lambda x: x[1], reverse=True)

plt.figure(figsize=(5, 4))
plt.title("2020年から2023年までの1年ごとの売り上げデータの棒グラフ")
plt.bar([xy[0] for xy in XY],[xy[1] for xy in XY])
plt.xlabel("年")
plt.ylabel("売り上げデータ")
plt.grid()
plt.show()
```
