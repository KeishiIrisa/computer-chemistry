```python
# 課題1

f = lambda x: x**2 + 2 * x + 1

for x in range(-3, 4, 2):
    print(f(x))
```

```python
# 課題2

g = lambda x: x**3 + 4*x**2 + 2*x + 2
for x in range(-2, 3, 1):
    print(g(x))
```

```python
# 課題3

from decimal import Decimal, getcontext
getcontext().prec = 4


for x in range(-15, 16, 1):
    x_decimal = Decimal(x*0.1)
    print(g(x_decimal))
```

```python
# 課題4
import matplotlib.pyplot as plt

X = []
Y = []

for x in range(-20, 21, 1):
    x_decimal = Decimal(x*0.5)
    X.append(x_decimal)
    Y.append(g(x_decimal))
    
plt.figure(figsize=(5, 4))
plt.title("Line graph of y = g(x) with 0.5 increments")
plt.plot(X, Y, label="y=g(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
# 課題5
import math

sigmoid = lambda x: 1 / (1 + math.exp(-x))

X = []
Y = []

for x in range(-50, 51, 1):

    x_decimal = Decimal(x*0.2)
    X.append(x_decimal)
    Y.append(sigmoid(x_decimal))
    
plt.figure(figsize=(5, 4))
plt.title("Line graph of the sigmoid function with 0.2 increments")
plt.plot(X, Y)
plt.xlabel("x")
plt.ylabel("y")
# plt.legend()
plt.grid()
plt.show()
```

```python
# 課題6

X = []
Y = []

for x in range(-100, 101, 1):
    x_decimal = Decimal(x*0.1)
    X.append(x_decimal)
    Y.append(math.exp(x_decimal))
    
plt.figure(figsize=(5, 4))
plt.title("Line graph of the exponential function with 0.1 increments")
plt.plot(X, Y)
plt.xlabel("x")
plt.ylabel("y")
# plt.legend()
plt.grid()
plt.show()
```

```python
# 課題7

# 通常のグラフ
X = []
Y = []

for x in range(5, 51, 1):
    x_decimal = Decimal(x*0.1)
    X.append(x_decimal)
    Y.append(math.log(x_decimal))
    
plt.figure(figsize=(5, 4))
plt.title("Graph of the natural logarithm function")
plt.plot(X, Y)
plt.xlabel("x")
plt.ylabel("y")
# plt.legend()
plt.grid()
plt.show()


# 対数グラフ  
plt.figure(figsize=(5, 4))
plt.title("Logarithmic graph of the natural logarithm function")
plt.plot(X, Y)
plt.xlabel("x(log scale)")
plt.ylabel("y(log scale)")
plt.xscale("log")
plt.yscale("log")
# plt.legend()
plt.grid()
plt.show()

```

```python
# 課題8
for x in range(-20, 21, 1):
    x_decimal = Decimal(x*0.5)
    X.append(x_decimal)
    Y.append(g(x_decimal))
    
X = [Decimal(x*0.1) for x in range(-30, 21, 1)]
Y = [g(x) for x in X]
    
plt.figure(figsize=(5, 4))
plt.title("Line graph of y = g(x) with 0.1 increments")
plt.plot(X, Y, label="y=g(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
# 課題9

pi = Decimal(math.pi)
sin = lambda x: math.sin(2*pi*x)
cos = lambda x: math.cos(2*pi*x)

X = [Decimal(x*0.01) for x in range(-200, 201, 1)]
sin_values = [sin(x) for x in X]
cos_values = [cos(x) for x in X]

plt.plot(X, sin_values, color="blue", label="sin(2πx)")
plt.plot(X, cos_values, color="red", label="cos(2πx)")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of sin(2πx) and cos(2πx)")

plt.legend()

plt.grid()
plt.show()
```
