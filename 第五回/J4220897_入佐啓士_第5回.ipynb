```python
# 課題1
import time
import math
import matplotlib.pyplot as plt

def g(x):
    return math.exp(x) -2*x -1

def bisection(f, a, b):
    tolerance = 1e-6
    max_iter = 100
    iter = 0
    while (b -a) / 2 > tolerance and iter < max_iter:
        c = (a + b) /2
        if f(c) == 0:
            break       
        elif f(c) * f(a) < 0:
            b =c
        else:
            a = c
        iter += 1
    return (a + b) /2

a = -3
b = 0

start_time = time.time()
solution = bisection(g, a, b)
end_time = time.time()
exe_time = end_time - start_time
print("方程式の解: ", round(solution, 5))
print("計算時間:", exe_time, "秒")

x = [i/10 for i in range(-30, 31)]
y = [g(i) for i in x]

plt.figure(figsize=(8, 6))
plt.plot(x, y, color="b")
plt.axhline(0, color="r", linestyle="--")

plt.xlabel("x")
plt.ylabel("y")
plt.title("e**x - 2x - 1")
plt.grid()
plt.show()


    
```

```python
# 課題2
def h(x):
    return math.tanh(x) - x**2 +1

def bisection(f, a, b):
    tolerance = 1e-6
    max_iter = 100
    iter = 0
    while (b -a) / 2 > tolerance and iter < max_iter:
        c = (a + b) /2
        if f(c) == 0:
            break       
        elif f(c) * f(a) < 0:
            b =c
        else:
            a = c
        iter += 1
    return (a + b) /2

a = -3
b = 0

start_time = time.time()
solution = bisection(h, a, b)
end_time = time.time()
exe_time = end_time - start_time
print("方程式の解: ", round(solution, 5))
print("計算時間:", exe_time, "秒")

x = [i/10 for i in range(-30, 31)]
y = [h(i) for i in x]

plt.figure(figsize=(8, 6))
plt.plot(x, y, color="b")
plt.axhline(0, color="r", linestyle="--")

plt.xlabel("x")
plt.ylabel("y")
plt.title("e**x - 2x - 1")
plt.grid()
plt.show()
```

```python
# 課題3
def df(x):
    return 2 *x

def newton(f, df, k):
    tolerance = 1e-6
    max_iter = 100
    iter = 0
    
    while abs(f(k)) > tolerance and iter < max_iter:
        k = k -f(k) / df(k)
        iter += 1
        
    return k

initial = 1.5

start_time = time.time()
solution = newton(g, df, initial)
end_time = time.time()
exe_time = end_time - start_time

print("方程式の解: ", round(solution, 5))
print("計算時間:", exe_time, "秒")
```

```python
# 課題4
def p(x):
    return math.exp(x) * math.cos(x) - 2*x

def newton(f, df, k):
    tolerance = 1e-6
    max_iter = 100
    iter = 0
    
    while abs(f(k)) > tolerance and iter < max_iter:
        k = k -f(k) / df(k)
        iter += 1
        
    return k

a = -1
b = 2 * math.pi

start_time = time.time()
solution = bisection(p, a, b)
end_time = time.time()
exe_time = end_time - start_time
print("方程式の解: ", round(solution, 5))
print("計算時間:", exe_time, "秒")

x = [i/10 for i in range(-30, 31)]
y = [p(i) for i in x]

plt.figure(figsize=(8, 6))
plt.plot(x, y, color="b")
plt.axhline(0, color="r", linestyle="--")

plt.xlabel("x")
plt.ylabel("y")
plt.title("e**xcosx - 2x")
plt.grid()
plt.show()
```

```python
# 課題5

def f(x):
    return x / (x**2 + 1)

def rectangular(f, a, b, h):
    n = int((b -a) / h)
    int_sum = 0
    
    for i in range(n):
        x = a + i*h
        int_sum += f(x) * h
    
    return int_sum

a = 1
b = 3
h_list = [0.1, 0.01, 0.001, 0.0001]

for h in h_list:
    start_time = time.time()
    integral = rectangular(f, a, b, h)
    end_time = time.time()
    exe_time = end_time - start_time
    print(f"積分値({h}): ", round(integral, 5))
    print("計算時間:", exe_time, "秒")
    
# hを小さくすると、解析解につかづくことが確認された。
```

```python
# 課題6

def f(x):
    return x / (x**2 + 1)

def trapezoidal(f, a, b, h):
    n = int((b -a) / h)
    int_sum = (f(a) + f(b)) / 2
    
    for i in range(1, n):
        x = a + i*h
        int_sum += f(x)
    
    int_sum *= h
    return int_sum

a = 1
b = 3
h_list = [0.1, 0.01, 0.001, 0.0001]

for h in h_list:
    start_time = time.time()
    integral = trapezoidal(f, a, b, h)
    end_time = time.time()
    exe_time = end_time - start_time
    print(f"積分値({h}): ", round(integral, 5))
    print("計算時間:", exe_time, "秒")
    
# 長方形近似よりも精度が高いことが確認できた。
```

```python
# 課題7
import random

def f(x):
    return math.sqrt(2*x + 1)

def monte_carlo(f, a, b, num_point):
    int_sum = 0
    for _ in range(num_point):
        x = random.uniform(a, b)
        int_sum += f(x)
    
    int_approx = (b - a)*(int_sum / num_point)
    return int_approx

a = 0
b = 4
num_point_list = [1000, 100000, 10000000]

for num_point in num_point_list:
    start_time = time.time()
    integral = monte_carlo(f, a, b, num_point)
    end_time = time.time()
    exe_time = end_time - start_time
    print(f"積分値({h}): ", round(integral, 5))
    print("計算時間:", exe_time, "秒")
    
# ランづうを増やすろ、解析解に近くなった。
```

```python
# 課題8
def f(x, y):
    return x**2 + y**3

def monte_carlo_double_integral(f, ax, bx, ay, by, num_point):
    int_sum = 0
    for _ in range(num_point):
        x = random.uniform(ax, bx)
        y = random.uniform(ay, by)
        int_sum += f(x, y)
    
    area = (bx - ax) * (by - ay)
    int_approx = area * (int_sum / num_point)
    return int_approx

ax = -1
bx = 1
ay = -1
by = 1
num_point = 10000000

start_time = time.time()
integral = monte_carlo_double_integral(f, ax, bx, ay, by, num_point)
end_time = time.time()
exe_time = end_time - start_time
print(f"積分値({num_point}): ", round(integral, 5))
print("計算時間:", exe_time, "秒")

# 積分値は解析解とかなり近い
```
